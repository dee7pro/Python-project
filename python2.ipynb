```python
#1. Running sum of 1 d array
class Solution:
    def runningSum(self, nums: [int]) -> list[int]:
        for i in range(1, len(nums)):
            nums[i] += nums[i - 1]
        return nums

ans = Solution()
print(ans.runningSum([1, 2, 3, 4]))
print(ans.runningSum([1, 1, 1, 1, 1]))
print(ans.runningSum([3, 1, 2, 10, 1]))
```

```python
#2. Shuffle the array 

from typing import List

class Solution:
    def shuffle(self, nums: List[int], n: int) -> List[int]:
        return[nums[i//2 + n*(i%2)] for i in range(2*n)] 
        
sol = Solution()
print(sol.shuffle([2,5,1,3,4,7], 3))     
print(sol.shuffle([1,2,3,4,4,3,2,1], 4)) 
print(sol.shuffle([1,1,2,2], 2))
```

```python
#3.Kids With the Greatest Number of Candies

from typing import List

class Solution:
    def kidsWithCandies(self, candies: List[int], extraCandies: int) -> List[bool]:
        maxCandies = max(candies)
        return [kid + extraCandies >= maxCandies for kid in candies]

sol = Solution()
candies1 = [2,3,5,1,3]
extra1 = 3
print(sol.kidsWithCandies(candies1, extra1))

candies2 = [4,2,1,1,2]
extra2 = 1
print(sol.kidsWithCandies(candies2, extra2))

candies3 = [12,1,12]
extra3 = 10
print(sol.kidsWithCandies(candies3, extra3))
```
